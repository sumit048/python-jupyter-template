### Exercise 3: Scatter Plot with Annotations

Scatter Plot

```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [5, 4, 3, 2, 1]

# Create a scatter plot
plt.scatter(x, y)

# Labeling the axes and the plot
plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatter Plot Example')

# Show the plot
plt.show()
```
